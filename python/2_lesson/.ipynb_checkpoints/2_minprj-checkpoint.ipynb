```python
import pandas as pd

bookings = pd.read_csv('2_bookings.csv', sep=";")
```

# 2

```python
def replace_space_with_star(name):
    new_name = name.replace(' ', '_')
    return new_name.lower()

bookings = bookings.rename(columns=replace_space_with_star)
bookings.head(1)
```

# 3

```python
bookings_top_5 = bookings.query('is_canceled == 0') \
                         .groupby('country') \
                         .agg({'is_canceled':'count'}) \
                         .sort_values('is_canceled', ascending=False) \
                         .head(5)
bookings_top_5
```

# 4

```python
bookings.groupby('hotel').agg({'stays_total_nights':'mean'})
```

# 5

```python
(bookings.assigned_room_type != bookings.reserved_room_type).sum()
```

# 6

```python
bookings.query('is_canceled == 0') \
        .groupby(['arrival_date_year','arrival_date_month'], as_index=False) \
        .agg({'stays_total_nights':'count'})
```

```python
# Чаще всего успешно оформляли бронь в 2016 - Октябрь
# Чаще всего успешно оформляли бронь в 2017 - Май
```

```python
bookings.query('hotel == "City Hotel" & is_canceled == 1') \
        .groupby(['arrival_date_year','arrival_date_month'], as_index=False) \
        .agg({'stays_total_nights':'count'})
```

```python
# Бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 
# 2015 - September
# 2016 - October
# 2017 - May
```

# 8

```python
bookings['total_kids'] = bookings.children + bookings.babies
bookings.groupby('hotel').agg({'total_kids':'mean'})
```

```python
# Resort Hotel
```

# 9

```python
bookings['has_kids'] = bookings.total_kids > 0
```

```python
# churn rate выше
bookings.query('has_kids == False & is_canceled == 1').shape[0] / bookings.query('has_kids == False').shape[0] * 100
```

```python
bookings.query('has_kids == True & is_canceled == 1').shape[0] / bookings.query('has_kids == True').shape[0] * 100
```
