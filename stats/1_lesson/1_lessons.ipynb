```python
# Визуализируйте динамику изменения объема мировых продаж (Global_Sales) по годам для игр Nintendo следующих жанров: 
# Fighting, Simulation, Platform, Racing, Sports. В какие моменты времени было продано больше всего игр жанра Sports?
```

```python
df.head()
```

```python
necc_genre = ["Fighting", "Simulation", "Platform", "Racing", "Sports"]
dff = df.query('Publisher == "Nintendo" & Genre in @necc_genre') \
        .groupby(["Year", "Genre"], as_index=False) \
        .agg({'Global_Sales':'sum'})
dff.head()
```

```python
plt.figure(figsize=(16,16))
sns.lineplot(x=dff['Year'], y=dff['Global_Sales'], hue=dff['Genre'])
```
